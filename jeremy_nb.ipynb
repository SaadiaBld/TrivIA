```python
import networkx as nx

# Créer un graphe vide
G = nx.Graph()

n_ext = 48
n_total = 69

# Création du cercle extérieur
for i in range(0, n_ext):
    G.add_node(i)

# Création des arrêtes extérieurs
for i in range(0, n_ext-1):
    G.add_edge(i, i + 1)

# La boucle est bouclée
G.add_edge(n_ext-1, 0)

# Création des chemins intérieurs
for i in range(n_ext, n_total):
    G.add_node(i)

centre = 53

# Branche 1
for i in range(48, 58):
    G.add_edge(i, i + 1)

# Branche 2
for i in range(59, 64):
    G.add_edge(i, i + 1)

# Branche 2 bis
for i in range(64, 69):
    G.add_edge(i, i + 1)


# Connexions spéciales à l'extérieur
G.add_edge(6, 59)
G.add_edge(18, 58)
G.add_edge(30, 68)
G.add_edge(42, 48)

# Connexions spéciales à l'intérieur
G.add_edge(53, 63)
G.add_edge(53, 64)
```

```python
positions = nx.spectral_layout(G)
```

```python
nx.draw_spectral(G, with_labels=True)
```

```python
import pandas as pd
import json

# Lire le fichier CSV
df = pd.read_csv('coor.csv')

# Transformer le DataFrame en dictionnaire en convertissant les types NumPy en types Python natifs
dict_resultat = {row['key'].item(): (row['x'].item(), row['y'].item()) for index, row in df.iterrows()}

# Convertir le dictionnaire en JSON
json_resultat = json.dumps(dict_resultat)

# Chemin du fichier où sauvegarder le JSON
chemin_fichier_json = 'mapping_coor.json'

# Sauvegarder le JSON dans un fichier
with open(chemin_fichier_json, 'w') as fichier:
    fichier.write(json_resultat)

print(f"JSON sauvegardé dans '{chemin_fichier_json}'")

```
